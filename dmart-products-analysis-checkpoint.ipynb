# Inter Quartile Range(IQR) Method

```python
Q1=df['Price'].quantile(0.25)                  # df.quantile(0.25) gives the 25th percentile of a column
Q3=df['Price'].quantile(0.75)                  # df.quantile(0.75) gives the 75th percentile of a column
Q1,Q3
```

```python
IQR = Q3-Q1                                   # IQR = Difference between 75th percentile and 25th percentile
IQR
```

```python
lower_limit = Q1 - 1.5*IQR                   # Lower limit and Upper limit to display the dataset without outliers.
upper_limit = Q3 + 1.5*IQR
lower_limit,upper_limit
```

```python
df_no = df[(df['Price'] > lower_limit) & (df['Price'] < upper_limit)]     # we display the datset without outliers
df_no
```

## So after applying IQR formula we removed outliers and we displayed the datset with no outliers in Price Column

```python
Q1=df['DiscountedPrice'].quantile(0.25)                     # df.quantile(0.25) gives the 25th percentile of a column
Q3=df['DiscountedPrice'].quantile(0.75)                     # df.quantile(0.75) gives the 75th percentile of a column
Q1,Q3
```

```python
IQR = Q3-Q1                                                 # IQR = Difference between 75th percentile and 25th percentile
IQR
```

```python
lower_limit1 = Q1 - 1.5*IQR                               # Lower limit and Upper limit to display the dataset without outliers.
upper_limit1 = Q3 + 1.5*IQR
lower_limit1,upper_limit1
```
